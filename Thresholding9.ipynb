```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread("doramon.jpg")
img = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
# simple
```

```python
thresh1 = cv.threshold(img,127,255,cv.THRESH_BINARY)[1]
thresh2 = cv.threshold(img,127,255,cv.THRESH_BINARY_INV)[1]
thresh3 = cv.threshold(img,127,255,cv.THRESH_TRUNC)[1]
thresh4 = cv.threshold(img,127,255,cv.THRESH_TOZERO)[1]
thresh5 = cv.threshold(img,127,255,cv.THRESH_TOZERO_INV)[1]
```

```python
# adaptive 
```

```python
athresh1 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,199,5)
athresh2 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,199,5)
```

```python
#otsu
```

```python
othresh = cv.threshold(img,120,255,cv.THRESH_BINARY+cv.THRESH_OTSU)[1]
```

```python
plt.figure(figsize=(15,10))
plt.subplot(4, 3, 1), plt.imshow(img, cmap='gray'), plt.title('Original Image')
plt.subplot(4, 3, 2), plt.imshow(thresh1, cmap='gray'), plt.title('Simple BINARY')
plt.subplot(4, 3, 3), plt.imshow(thresh2, cmap='gray'), plt.title('Simple BINARY_INV')
plt.subplot(4, 3, 4), plt.imshow(thresh3, cmap='gray'), plt.title('Simple TRUNC')
plt.subplot(4, 3, 5), plt.imshow(thresh4, cmap='gray'), plt.title('Simple TOZERO')
plt.subplot(4, 3, 6), plt.imshow(thresh5, cmap='gray'), plt.title('Simple TOZERO_INV')
plt.subplot(4, 3, 7), plt.imshow(athresh1, cmap='gray'), plt.title('Adaptive (mean)')
plt.subplot(4, 3, 9), plt.imshow(athresh2, cmap='gray'), plt.title('Adaptive (gaussian mean)')
plt.subplot(4,3,11), plt.imshow(othresh, cmap='gray'), plt.title('Otsu Thresholding')

plt.tight_layout()
plt.show()
```
